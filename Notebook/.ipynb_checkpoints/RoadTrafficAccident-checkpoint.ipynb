## Data Transformation

```python
#Remove special charaters in columns
data.columns = data.columns.str.replace('[#,@,&,<,>]', '')
```

```python
print("Before removing duplicates --Rows",data.shape[0])
data.drop_duplicates(keep="first",inplace=True)
print("After removing duplicates --Rows",data.shape[0])
```

#### Handling Missing values :
     
     1.Forward fill first and followed by backward fill for less than 1000 missing records.
     2.Datawig.simpleImputer method for columns more than 1000 missing records with precison threshold of 80%.
     3.Mode filling applied for rest of the missing records.
    

```python
data.isnull().sum()
```

#### Observation: Displays the number of missing values for each feature.

    1. 16 features are having missing values.

    2. Column with highest missing values is Defect_of_Vehicle.
             
    3. Column with more than 1000 missing values['Service_year_of_vehicle',                                        
                                                 'Defect_of_vehicle', 
                                                 'Work_of_casuality',

#### Applying forward & backward fill for columns less than 1000 missing records

```python
#Removing the features having more than 1000 missing values  and label encoding columns
dropCol=['Fitness_of_casuality','Defect_of_vehicle', 'Service_year_of_vehicle','Work_of_casuality' ]
data4=data.drop(dropCol,axis=1)
                 
for col in data4.columns:
    if(data4[col].isnull().sum()!=0):
        data4[col]=data4[col].ffill(axis=0)
for col in data4.columns:
    if(data4[col].isnull().sum()!=0):
        data4[col]=data4[col].bfill(axis=0)
data4=data4.join(data[dropCol])
```

```python
data4.isna().sum()
```

#### Applying Datawig(Amazon's library)Simple imputer to fill more than 1000 missing records.It uses deep learning model to fill.

```python
df = datawig.SimpleImputer.complete(data4,
                                    precision_threshold=0.8,
                                    inplace=False,
                                    hpo=False,
                                    verbose=0,
                                    num_epochs=100,
                                    iterations=1,
                                    output_path='.'
                                   )
```

#### Filling with most frequent values using sklearn simpleImputer for the remaining unfilled missing columns

```python

value = df.values
 # defining the imputer
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
 # transform the dataset
transformed_values = imputer.fit_transform(value)
 
data4 = pd.DataFrame(transformed_values, columns =data4.columns )
```

### Encoding the data 

```python
data4.drop(['Accident_severity'],axis=1,inplace=True)
```

```python
data4=data4.join(df['Accident_severity'])
df = data4.apply(LabelEncoder().fit_transform)
```

#### Splitting the data 

```python
#Split the data with train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.33, random_state=42)
```

## Feature Selection Method - Chi square 

```python
def select_features(X_train, y_train, X_test):
	fs = SelectKBest(score_func=chi2, k='all')
	fs.fit(X_train, y_train)
	X_train_fs = fs.transform(X_train)
	X_test_fs = fs.transform(X_test)
	return X_train_fs, X_test_fs, fs
X_train_fs, X_test_fs, fs = select_features(X_train, Y_train, X_test)
```

```python
#Feature Ranking 
X_traindf = pd.DataFrame(data = X_train) 
importances = fs.scores_

indices = np.argsort(importances)[::-1]
indices
copyIndicies = indices
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]),'-',X_traindf.columns[indices[f]])

# plot the scores
sns.barplot(y = fs.scores_,x = X_traindf.columns)
plt.xticks(rotation=90)
plt.title('Chi-Squared - Feature Importance Plot')    
```

```python
# Get the dataset with only 94 features after FSM
featurecol = X_traindf.columns[indices[0:19]]
df2 = df[featurecol]  
X=df2.iloc[:,:-1]
Y=df.iloc[:,-1]
```

### 18 Features are selected accoridng to their chi-square values.

```python
#Split the data with train and test data after FSM(Feature Selection Method
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

### Upsampling 

```python
# upsampling using smote

counter = Counter(Y_train)

print("=============================")

for k,v in counter.items():
    per = 100*v/len(Y_train)
    print(f"Class= {k}, n={v} ({per:.2f}%)")

oversample = SMOTE()
X_train, Y_train = oversample.fit_resample(X_train, Y_train)

counter = Counter(Y_train)

print("=============================")

for k,v in counter.items():
    per = 100*v/len(Y_train)
    print(f"Class= {k}, n={v} ({per:.2f}%)")

print("=============================")

print("Upsampled data shape: ", X_train.shape, Y_train.shape)
```

```python
def modelling(X_train, y_train, X_test, y_test, **kwargs):
    scores = {}
    models = []
    if 'xgb' in kwargs.keys() and kwargs['xgb']:
        xgb = XGBClassifier()
        xgb.fit(X_train._get_numeric_data(), np.ravel(y_train, order='C'))
        y_pred = xgb.predict(X_test._get_numeric_data())
        scores['xgb']= [f1_score(y_test, y_pred, average= 'weighted')]
#         scores['xgb']['roc_auc'] = roc_auc_score(y_test, y_pred)

    if 'rf' in kwargs.keys() and kwargs['rf']:
        rf = RandomForestClassifier(n_estimators=200)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_test)
        scores['rf']= [f1_score(y_test, y_pred, average= 'weighted')]
#         scores['rf']['roc_auc'] = roc_auc_score(y_test, y_pred)
        models.append(rf)

    if 'extree' in kwargs.keys() and kwargs['extree']:
        extree = ExtraTreesClassifier()
        extree.fit(X_train, y_train)
        y_pred = extree.predict(X_test)
        scores['extree'] = [f1_score(y_test, y_pred, average= 'weighted')]
#         scores['extree']['roc_auc'] = roc_auc_score(y_test, y_pred)
        models.append(extree)
        
    return scores
```

```python
modelling(X_train,Y_train, X_test, Y_test, xgb=True, rf=True, extree=True)
```

```python
rf.get_params()
```

### Random Forest Model Building & Hyperparameter Tuning

```python
gkf = KFold(n_splits=3, shuffle=True, random_state=42).split(X=X_train, y=Y_train)


params = {'bootstrap': [True, False],
 'max_depth': [10, 20,100],
 'max_features': ['auto'],
 'min_samples_leaf': [1],
 'min_samples_split': [2],
 'n_estimators': [400,600]}

rf_estimator = RandomForestClassifier()

gsearch = GridSearchCV(
    estimator= rf_estimator,
    param_grid= params,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf,
    verbose=3,
)

rf_model = gsearch.fit(X=X_train, y=Y_train)
(gsearch.best_params_, gsearch.best_score_)

```

```python
gkf = KFold(n_splits=3, shuffle=True, random_state=42).split(X=X_train, y=Y_train)

# A parameter grid for RandomForest Estimator
params = {'bootstrap': [False],
 'max_depth': [100],
 'max_features': ['auto'],
 'min_samples_leaf': [1],
 'min_samples_split': [2],
 'n_estimators': [600],
 }

rf_estimator = RandomForestClassifier()

gsearch = GridSearchCV(
    estimator= rf_estimator,
    param_grid= params,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf,
    verbose=3,
)

rf_model = gsearch.fit(X=X_train, y=Y_train)
(gsearch.best_params_, gsearch.best_score_)

```

```python
bestModel=gsearch.best_estimator_
```

```python
bestModel.fit(X_train, Y_train)
y_pred_tuned =bestModel.predict(X_test)
```
